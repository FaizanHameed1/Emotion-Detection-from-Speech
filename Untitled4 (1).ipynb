```python
#Data augmantation
#Some important techniques of data augmentation are "time stretching", "noise addition","pitch shifting" and "time shifting" of data. 
```

```python
#noise addition with normal distribution of noise(mean=0,standard deviation=1)
#We will genrate distribution with numpy normal method and then add it to the orignal signal
#Permissible noise factor value = x > 0.004
noisy=[]
for i in df.directory:
  data, sr = librosa.load(i, sr=44100)#array of data and learning rate
  dist = 0.035*np.random.uniform()*np.amax(data)#distribution
  data = data + dist*np.random.normal(size=data.shape[0])#adding noise distribution to the signal
  noisy.append(data)

noisy[0:5]
```

```python
#time stretching


time_stret=[]
for i in df.directory:
  rate=0.8
  data, sr = librosa.load(i, sr=44100)
  stretch=librosa.effects.time_stretch(data, rate)
  time_stret.append(stretch)

time_stret[0:5]
```

```python
time_stret[0:5]
```

```python
#time_shifting
time_shift=[]
for i in df.directory:
  data, sr = librosa.load(i, sr=44100)
  shift_range = int(np.random.uniform(low=-5, high = 5)*1000)
  t_shift= np.roll(data, shift_range)
  time_shift.append(t_shift)
time_shift[0:5]
```

```python
#pitch_shifting
pitch_shift=[]
for i in df.directory:
  pitch_factor=0.7
  sampling_rate=44100
  data, sr = librosa.load(i, sr=44100)
  p_shift=librosa.effects.pitch_shift(data, sampling_rate, pitch_factor)
  pitch_shift.append(p_shift)
pitch_shift[0:5]
```
